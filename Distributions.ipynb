```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import warnings
warnings.filterwarnings("ignore")
```

```python
#normal distributon/gussain distribution
mu,sigma=0,0.1
s=np.random.normal(mu,sigma,size=10000)
print("Count of DataPoints :",len(s))
s
```

```python
plt.hist(s, 30, density = True)
plt.show()
```

```python
sns.distplot(s)
plt.show()
```

```python
#uniform dist>all the outcomes has equal prob to occur in range
u=np.random.uniform(0,1,1000)
plt.hist(u)
```

```python
#log normal dist>> whose logarithm is normally distributed>> right skewed
l = np.random.lognormal(mean = 0, sigma=0.5, size = 1000)
l = l[np.where(abs(l) < 5)]
plt.hist(l, bins=100)
```

```python
#log normal--> normal>> log transformation
#Converting lognormal distribution to normal distribution
sns.distplot(np.log(l))
```

```python
#Square transformation
sns.distplot(np.sqrt(l))
```

```python
#Chi -square distribution
c=np.random.chisquare(df=1,size=10000)
c= c[np.where(abs(c) < 10)]
plt.hist(c, bins = 1000)
plt.show()
```

```python
#t distribstion t dist with df 1>> tail is thicker/larger
t=np.random.standard_t(df=1,size=10000)
t=t[np.where(abs(t)<10)]
plt.hist(t,bins = 1000)
plt.show()
```

```python
#t dist with df 10>> tail is more thinner/smaller, approxx normal dist
t=np.random.standard_t(df=31,size=10000)
t=t[np.where(abs(t)<10)]
plt.hist(t,bins = 1000)
plt.show()
```

```python
#F distribution
#dfnum is df for numerator, dfden is df for den
f=np.random.f(dfnum = 1, dfden=1,size=10000)
f=f[np.where(abs(f)<10)]
plt.hist(f,bins=1000)
plt.show()
```

```python
f=np.random.f(dfnum = 3, dfden=10,size=10000)
f=f[np.where(abs(f)<10)]
plt.hist(f,bins=1000)
plt.show()
```

```python
#bernoulli dist>>> binomial dist with 1 trial, n bernouli trial is binomial dist
b = np.random.binomial(n=1, p=0.5, size = 1000)
plt.hist(b)
plt.show()
```

```python
#binomial dist

b = np.random.binomial(n=100, p=0.1, size = 1000)
plt.hist(b, bins = 20)
plt.show()
```

```python
#poisson dist
p = np.random.poisson(lam = 5,size = 1000)
plt.hist(p, bins = 20)
plt.show()
```
